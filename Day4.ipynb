```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import scale
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
iris = load_iris()
iris_columns = iris.feature_names

iris= pd.DataFrame(data = iris.data, columns= iris_columns)
iris_only_2_features = iris[['petal length (cm)' , 'petal width (cm)']]
```

```python
fig, axs = plt.subplots(figsize = (7,3))

plt.scatter(x= iris_only_2_features.iloc[:,0], 
            y = iris_only_2_features.iloc[:,1], 
            cmap='jet' )
plt.xlabel(iris_only_2_features.columns[0])
plt.ylabel(iris_only_2_features.columns[1])
plt.title("iris_only_2_features")

%matplotlib inline
```

```python
km_model = KMeans(n_clusters=3)
km_model.fit(iris_only_2_features)
iris_only_2_features['cluster'] = km_model.labels_
```

```python
fig, axs = plt.subplots(2,1,figsize = (7,5))

axs[0].scatter(x= iris_only_2_features.iloc[:,0], 
            y = iris_only_2_features.iloc[:,1], 
            cmap='jet' )

axs[1].scatter(iris_only_2_features.iloc[:,0], 
            iris_only_2_features.iloc[:,1], 
            c=iris_only_2_features.iloc[:,2])
```

```python
from matplotlib.image import imread

image = imread('./images.jpeg')
print(image.shape)
plt.imshow(image)
```

```python
X = image.reshape(-1,3)
kmeans = KMeans(n_clusters= 3 , random_state= 1).fit(X) # 비슷한 색상으로 군집 BUT 우선 1차원으로 
```

```python
segmented_img = kmeans.cluster_centers_[kmeans.labels_]
segmented_img = segmented_img.reshape(image.shape)
```
